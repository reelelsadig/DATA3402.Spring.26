1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.value = 0
    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("Error: Counter has reached its maximum value.")
    def reset(self):
        self.value = 0 
    def get_value(self):
        return self.value
```

```python
c = counter(3)
print("Initial value:", c.get_value())
c.increment()
print("After 1 increment:", c.get_value())
c.increment()
c.increment()
print("After 2 increments:", c.get_value())
c.increment()
c.increment()
print("After 3 increments:", c.get_value())
c.reset()
print("After reset:", c.get_value())
```

```python
class Counter:
    def __init__(self, max_value):
        self.__max_value = max_value
        self.__value = 0
    def increment(self):
        if self.__value < self.__max_value:
            self.__value += 1
        else:
            print("Error: Counter has reached its maximum value.")
    def reset(self):
        self.__value = 0 
    def get_value(self):
        return self.__value
    def get_max_value(self):
        return self.__max_value
    def is_at_max(self):
        return self.__value == self.__max_value
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
import math

class Circle:
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self.__radius
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
import math

class Shape:
    def __init__(self, x, y):
        self._x = x
        self._y = y

    def contains(self, px, py):
        raise NotImplementedError("Subclasses must implement this method.")

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y) # Passes x, y to the Shape base class
        self.__length = length
        self.__width = width

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def contains(self, px, py):
        return (self._x <= px <= self._x + self.__length) and \
               (self._y <= py <= self._y + self.__width)

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius

    def area(self):
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def contains(self, px, py):
        distance = math.sqrt((px - self._x)**2 + (py - self._y)**2)
        return distance <= self.__radius

```

```python
r = Rectangle(10, 5, 2, 3)
c = Circle(5, 1, 2)

print(f"Rectangle contains (5, 5): {r.contains(5, 5)}")   # Inside: True
print(f"Rectangle contains (20, 20): {r.contains(20, 20)}") # Outside: False
print(f"Circle contains (2, 2): {c.contains(2, 2)}")      # Inside: True

            
        
```

```python

import math

class Shape:
    def __init__(self, x, y):
        self._x = x
        self._y = y

    def parameter_points(self, num_points=16):
        raise NotImplementedError("Subclasses must implement this")

    def contains(self, px, py):
        raise NotImplementedError("Subclasses must implement this")

    # --- Question 9: This is the missing method! ---
    def overlaps(self, other):
        # Check if any of our perimeter points are inside the other shape
        for px, py in self.parameter_points(16):
            if other.contains(px, py):
                return True
        
        # Check if any of the other shape's points are inside us
        for px, py in other.parameter_points(16):
            if self.contains(px, py):
                return True
                
        return False

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self._radius = radius

    def parameter_points(self, num_points=16):
        num_points = min(num_points, 16)
        points = []
        for i in range(num_points):
            angle = (2 * math.pi * i) / num_points
            px = self._x + self._radius * math.cos(angle)
            py = self._y + self._radius * math.sin(angle)
            points.append((px, py))
        return points

    def contains(self, px, py):
        distance = math.sqrt((px - self._x)**2 + (py - self._y)**2)
        return distance <= self._radius

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self._length = length
        self._width = width

    def parameter_points(self, num_points=16):
        num_points = min(num_points, 16)
        points = []
        # Simplified perimeter points for testing
        for i in range(num_points):
            t = i / num_points
            if t < 0.25: px, py = self._x + (t*4)*self._length, self._y
            elif t < 0.5: px, py = self._x + self._length, self._y + (t-0.25)*4*self._width
            elif t < 0.75: px, py = self._x + self._length - (t-0.5)*4*self._length, self._y + self._width
            else: px, py = self._x, self._y + self._width - (t-0.75)*4*self._width
            points.append((px, py))
        return points

    def contains(self, px, py):
        return (self._x <= px <= self._x + self._length) and \
               (self._y <= py <= self._y + self._width)
      
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
