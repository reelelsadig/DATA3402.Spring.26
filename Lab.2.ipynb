*Exercise 1:* Write a function that creates an **n×n matrix** (a list of lists) representing the state of a tic‑tac‑toe game.

Use the integers:

- `0` = empty
- `1` = `"X"`
- `2` = `"O"`


````python
# Write your solution here
def create_board(n):
    board = []
    for i in range(n):
        row = []
        for j in range(n):
            row.append(0)
        board.append(row)
    return board
````

*Exercise 3:* Modify Exercise 2 so that it takes a matrix in the format from Exercise 1 and draws a tic‑tac‑toe board with `"X"`s and `"O"`s.

````python
# Write your solution 
def draw_board(board):
    n = len(board)

    for i in range(n):
        print(" ---" * n)
        row_str = ""
        for j in range(n):
            if board[i][j] == 1:
                mark = "X"
            elif board[i][j] == 2:
                mark = "O"
            else:
                mark = " "
            row_str += "| " + mark + " "
        row_str += "|"
        print(row_str)
    print(" ---" * n)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
# Write your solution here
def check_winner(board):
    n = len(board)

    for row in board:
        if row.count(row[0]) == n and row[0] != 0:
            return row[0]

    for col in range(n):
        first = board[0][col]
        win = True
        for row in range(n):
            if board[row][col] != first:
                win = False
        if win and first != 0:
            return first

    first = board[0][0]
    win = True
    for i in range(n):
        if board[i][i] != first:
            win = False
    if win and first != 0:
        return first

    first = board[0][n-1]
    win = True
    for i in range(n):
        if board[i][n-1-i] != first:
            win = False
    if win and first != 0:
        return first

    for row in board:
        if 0 in row:
            return -1

    return 0
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
# Write your solution here
def make_move(board, player, row, col):
    if board[row][col] == 0:
        board[row][col] = player
        return True
    return False
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
# Write your solution here
def move_from_string(board, player, loc):
    loc = loc.upper()
    row_letter = loc[0]
    col_number = int(loc[1:]) - 1

    letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    row = letters.index(row_letter)

    return make_move(board, player, row, col_number)
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
# Write yourrr solution here
def play_game(n):
    board = create_board(n)
    player = 1

    while True:
        draw_board_labeled(board)
        player_turn(board, player)

        result = check_winner(board)

        if result == 1:
            draw_board_labeled(board)
            print("Player 1 (X) wins!")
            break
        elif result == 2:
            draw_board_labeled(board)
            print("Player 2 (O) wins!")
            break
        elif result == 0:
            draw_board_labeled(board)
            print("Draw!")
            break

        # switch player
        if player == 1:
            player = 2
        else:
            player = 1
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
# Write your solution here
import random

class Board:
    def __init__(self, n):
        self.n = n
        self.grid = [[" "] * n for _ in range(n)]

    def moves(self):
        return [(r, c) for r in range(self.n) for c in range(self.n)
                if self.grid[r][c] == " "]

    def play(self, move, player):
        r, c = move
        self.grid[r][c] = player

    def undo(self, move):
        r, c = move
        self.grid[r][c] = " "

    def win(self, p):
        n = self.n
        g = self.grid
        for i in range(n):
            if all(g[i][j] == p for j in range(n)): return True
            if all(g[j][i] == p for j in range(n)): return True
        if all(g[i][i] == p for i in range(n)): return True
        if all(g[i][n-1-i] == p for i in range(n)): return True
        return False

    def full(self):
        return all(self.grid[r][c] != " "
                   for r in range(self.n) for c in range(self.n))


def simple_ai(board, player):
    opponent = "O" if player == "X" else "X"

    # try to win
    for move in board.moves():
        board.play(move, player)
        if board.win(player):
            board.undo(move)
            return move
        board.undo(move)

    # block opponent
    for move in board.moves():
        board.play(move, opponent)
        if board.win(opponent):
            board.undo(move)
            return move
        board.undo(move)

    return random.choice(board.moves())
````

````python
# Test your solution here
b = Board(3)
print(simple_ai(b, "X"))
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write your solution here
import math

def minimax(board, player, depth, max_depth):
    opponent = "O" if player == "X" else "X"

    if board.win("X"): return -1
    if board.win("O"): return 1
    if board.full() or depth == max_depth: return 0

    scores = []
    for move in board.moves():
        board.play(move, player)
        score = minimax(board, opponent, depth + 1, max_depth)
        board.undo(move)
        scores.append(score)

    return max(scores) if player == "O" else min(scores)


def smart_ai(board, player, max_depth):
    best_move = None
    best_score = -math.inf if player == "O" else math.inf

    for move in board.moves():
        board.play(move, player)
        score = minimax(
            board,
            "O" if player == "X" else "X",
            1,
            max_depth
        )
        board.undo(move)

        if player == "O" and score > best_score:
            best_score = score
            best_move = move
        if player == "X" and score < best_score:
            best_score = score
            best_move = move

    return best_move
````

````python
# Test your solution here
b = Board(3)
print(smart_ai(b, "X", max_depth=3))
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
# Write your solution here
def play_game(n, ai1, ai2, depth):
    board = Board(n)
    players = [("X", ai1), ("O", ai2)]
    turn = 0

    while True:
        p, ai = players[turn % 2]
        move = ai(board, p) if ai == simple_ai else ai(board, p, depth)
        board.play(move, p)

        if board.win(p):
            return p
        if board.full():
            return "Draw"

        turn += 1
````

````python
# Test your solution here
for n in [3, 4, 5]:
    smart_wins = 0
    games = 10

    for i in range(games):
        if i % 2 == 0:
            winner = play_game(n, simple_ai, smart_ai, depth=3)
            if winner == "O":
                smart_wins += 1
        else:
            winner = play_game(n, smart_ai, simple_ai, depth=3)
            if winner == "X":
                smart_wins += 1

    print(f"{n}x{n} smart AI win rate:", smart_wins / games)
````
