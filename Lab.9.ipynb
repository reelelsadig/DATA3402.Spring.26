## Exercise 2

Following the original paper (see lab 7), make a comparison of the performance (using ROC curves and AUC) between models trained with raw, features, and raw+features data.

```python
RawNames = ["l_1_pT", "l_1_eta","l_1_phi",
            "l_2_pT", "l_2_eta", "l_2_phi",
            "MET", "MET_phi", "MET_rel", "axial_MET"]

FeatureNames = ["M_R", "M_TR_2", "R", "MT2",
                "S_R", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_curve, auc

def run_model(features):
    Xtr = Train_Sample[features].values
    Xte = Test_Sample[features].values

    model = MLPClassifier(hidden_layer_sizes=(32,16), max_iter=30)
    model.fit(Xtr, y_Train)

    y_pred = model.predict_proba(Xte)[:,1]

    fpr, tpr, _ = roc_curve(y_Test, y_pred)
    roc_auc = auc(fpr, tpr)

    return fpr, tpr, roc_auc
```

```python
print(df.columns)
```

```python
all_cols = df.drop(columns=["signal"]).columns

RawNames = all_cols[:10]      # first half
FeatureNames = all_cols[10:]  # second half
```

```python
all_features = list(RawNames) + list(FeatureNames)
```

```python
# Raw only
fpr_raw, tpr_raw, auc_raw = run_model(RawNames)

# Features only
fpr_feat, tpr_feat, auc_feat = run_model(FeatureNames)

# Combined
all_features = list(RawNames) + list(FeatureNames)
fpr_all, tpr_all, auc_all = run_model(all_features)
```

```python
import matplotlib.pyplot as plt

plt.plot(fpr_raw, tpr_raw, label=f"Raw (AUC={auc_raw:.2f})")
plt.plot(fpr_feat, tpr_feat, label=f"Features (AUC={auc_feat:.2f})")
plt.plot(fpr_all, tpr_all, label=f"All (AUC={auc_all:.2f})")

plt.legend()
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Comparison")
plt.show()
```

## Exercise 3

Design and implement at least 3 different DNN models. Train them and compare performance. You may try different architectures, loss functions, and optimizers to see if there is an effect.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_curve, auc

def run_model_layers(layers):
    model = MLPClassifier(hidden_layer_sizes=layers, max_iter=50)
    model.fit(X_Train, y_Train)

    y_pred = model.predict_proba(X_Test)[:,1]
    fpr, tpr, _ = roc_curve(y_Test, y_pred)
    roc_auc = auc(fpr, tpr)

    return fpr, tpr, roc_auc
```

```python
# Model 1 (simple)
fpr1, tpr1, auc1 = run_model_layers((32,))

# Model 2 (medium)
fpr2, tpr2, auc2 = run_model_layers((64,32))

# Model 3 (deep)
fpr3, tpr3, auc3 = run_model_layers((128,64,32))
```

```python
import matplotlib.pyplot as plt

plt.plot(fpr1, tpr1, label=f"(32) AUC={auc1:.2f}")
plt.plot(fpr2, tpr2, label=f"(64,32) AUC={auc2:.2f}")
plt.plot(fpr3, tpr3, label=f"(128,64,32) AUC={auc3:.2f}")

plt.legend()
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("Model Architecture Comparison")
plt.show()
```
